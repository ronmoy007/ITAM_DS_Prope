
Aproximemos a la función $\frac{1}{x}$ en el intervalo $[1,2]$ con polinomios de Taylor de orden $n$ con $n \in \{0,1,2\}$ con centro en $x_0=1.5$. Los polinomios de Taylor son:
$$P_0(x) = f(x_0) = \frac{2}{3} \quad \text{(constante)}$$
$$P_1(x) = f(x_0) + f^{(1)}(x_0)(x-x_0) = \frac{2}{3} - \frac{1}{x_0^2}(x-x_0) \quad \text{(lineal)}$$
$$P_2(x) = f(x_0) + f^{(1)}(x_0)(x-x_0) + \frac{f^{(2)}(x_0)(x-x_0)^2}{2} = \frac{2}{3} - \frac{1}{x_0^2}(x-x_0) + \frac{2}{x_0^3}(x-x_0)^2 \quad \text{(cuadrático)}$$
Ejercicio: graficar la función y los polinomios constante, lineal y cuadrático en una sola gráfica con matplotlib en el intervalo [1,2]. ¿Cuánto es la aproximación de los polinomios en x=1.9? Calcula el error relativo de tus aproximaciones.

```python
#import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
import mjmaLibMat as LM

#Declaramos las funciones que usaremos para obtener los polinomios de Taylor
def funcOrigen(nbrX):
    return 1/nbrX

def funcPolinomioTaylor0(nbrX,nbrXCero):
    nbrResultado=funcOrigen(nbrXCero)
    return nbrResultado

def funcPolinomioTaylor1(nbrX,nbrXCero):
    nbrResultado=funcPolinomioTaylor0(nbrX,nbrXCero)
    nbrResultado=nbrResultado+(-1/(nbrXCero**2))*(nbrX-nbrXCero)
    return nbrResultado
    
def funcPolinomioTaylor2(nbrX,nbrXCero):
    nbrResultado=funcPolinomioTaylor1(nbrX,nbrXCero)
    nbrResultado=nbrResultado+((2)/(nbrXCero**3))*(((nbrX-nbrXCero)**2)/2)
    return nbrResultado

def funcObtenerArregloXRangoXPolinomio(funcPolinomioTaylorN, arrIntervalo, nbrXCero):
    
    #Declaramos variables
    arrResultados = []
    nbrResultadoFuncion=0
    i=1
    
    '''
    Este for tomará el intervalo que recibimos como parámetro,
    y generará un array con los valores que devuelva el polinomio correspondiente en ese intervalo
    '''
    for i in arrIntervalo:
        nbrResultadoFuncion=funcPolinomioTaylorN(i,nbrXCero)
        arrResultados.append(nbrResultadoFuncion)
        
    return arrResultados


#Inicializamos los parámetros para calcular los valores de los diferentes polinomios
#Modifiqué el rango de la función para que las gráficas fuesen más claras
nbrXCero=1.5
nbrRangoInicio=.01
nbrRangoFin=2
nbrIntervalo=.01

nbrPuntoEval=1.9
nbrResultadoRealPuntoEval=0
nbrValorAproxEnT0=0
nbrValorAproxEnT1=0
nbrValorAproxEnT2=0
nbrErrorRelatT0=0
nbrErrorRelatT1=0
nbrErrorRelatT2=0

#Arreglos que usaremos
arrIntervalo=np.arange(nbrRangoInicio, nbrRangoFin + nbrIntervalo, nbrIntervalo)
arrFuncOrigen=np.arange(0)
arrTaylor0=np.arange(0)
arrTaylor1=np.arange(0)
arrTaylor2=np.arange(0)

print('Función origen')
for i in arrIntervalo:
    if i != 0:
        nbrResultado=funcOrigen(i)
        arrFuncOrigen=np.append(arrFuncOrigen,nbrResultado)
plt.plot(arrIntervalo,arrFuncOrigen)
plt.grid(True)
plt.show()

print('PolinomioTaylor0')
arrTaylor0=funcObtenerArregloXRangoXPolinomio(funcPolinomioTaylor0,arrIntervalo,nbrXCero)
plt.plot(arrIntervalo,arrTaylor0)
plt.grid(True)
plt.show()

print('PolinomioTaylor1')
arrTaylor1=funcObtenerArregloXRangoXPolinomio(funcPolinomioTaylor1,arrIntervalo,nbrXCero)
plt.plot(arrIntervalo,arrTaylor1)
plt.grid(True)
plt.show()

print('PolinomioTaylor2')
arrTaylor2=funcObtenerArregloXRangoXPolinomio(funcPolinomioTaylor2,arrIntervalo,nbrXCero)
plt.plot(arrIntervalo,arrTaylor2)
plt.grid(True)
plt.show()

nbrResultadoRealPuntoEval=funcOrigen(nbrPuntoEval)

nbrValorAproxEnT0=funcPolinomioTaylor0(nbrPuntoEval,nbrXCero)
nbrValorAproxEnT1=funcPolinomioTaylor1(nbrPuntoEval,nbrXCero)
nbrValorAproxEnT2=funcPolinomioTaylor2(nbrPuntoEval,nbrXCero)

print('nbrPuntoEval: ', nbrPuntoEval)
print('nbrResultadoRealPuntoEval: ',nbrResultadoRealPuntoEval)
print('nbrValorAproxEnT0: ', nbrValorAproxEnT0)
print('nbrValorAproxEnT1: ', nbrValorAproxEnT1)
print('nbrValorAproxEnT2: ', nbrValorAproxEnT2)
print('')

nbrErrorRelatT0=LM.CalcErrorRelat(nbrValorAproxEnT0,nbrResultadoRealPuntoEval)
nbrErrorRelatT1=LM.CalcErrorRelat(nbrValorAproxEnT1,nbrResultadoRealPuntoEval)
nbrErrorRelatT2=LM.CalcErrorRelat(nbrValorAproxEnT2,nbrResultadoRealPuntoEval)

print('nbrErrorRelatT0: ', nbrErrorRelatT0)
print('nbrErrorRelatT1: ', nbrErrorRelatT1)
print('nbrErrorRelatT2: ', nbrErrorRelatT2)
print('')

print('Los 3 polinomios')
plt.plot(arrIntervalo,arrTaylor0)
plt.plot(arrIntervalo,arrTaylor1)
plt.plot(arrIntervalo,arrTaylor2)
plt.legend(('Tay0 en 1.9: '+ str(nbrValorAproxEnT0) ,'Tay1 en 1.9: '+ str(nbrValorAproxEnT1) ,'Tay2 en 1.9: '+ str(nbrValorAproxEnT2) ),loc = 0)
plt.grid(True)
plt.show()


```
