Ejercicio: usar esta matriz de rotación para rotar un ángulo de $\frac{\pi}{3}$ en sentido contrario a las manecillas del reloj al vector $\left(\begin{array}{c}
-3\\
4\\
\end{array}
\right)$.

```python
#Importamos
import numpy as np
import math
import matplotlib.pyplot as plt
import mjmaLibMat as LM
import pprint
import scipy
import scipy.linalg 
```

```python
#Valor del ángulo para realizar la rotación
nbrAngulo=(math.pi/3)

#Matriz de rotación
mtrxRot=np.array([[math.cos(nbrAngulo),-math.sin(nbrAngulo)],
                  [math.sin(nbrAngulo), math.cos(nbrAngulo)]])

#Vector a rotar
arrV=np.array([-3,4])

#para probar
#arrV=np.array([2,2])

print('Vector original: ', arrV)
print('Norma vector original: ', np.linalg.norm(arrV))

#Rotamos el vector mediante la matriz de rotación
arrVRslt=mtrxRot@arrV

print('Vector rotado: ', arrVRslt)
print('Norma vector rotado: ', np.linalg.norm(arrVRslt))

#Pintamos la línea del vector original
x=[0,arrV[0]]
y=[0,arrV[1]]
plt.plot(x,y)

#Pintamos la línea del vector rotado
x=[0,arrVRslt[0]]
y=[0,arrVRslt[1]]
plt.plot(x,y)

plt.grid(True)
plt.legend(('Vector original','Vector rotado'), loc=0)

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.gca().set_aspect('equal')

plt.show()
```

Ejercicio: resolver el ejercicio 3) anterior con la factorización QR del paquete numpy de la matriz A y verificar resultado

$$ A = \left[\begin{array}{ccc}
1 & 4& 1\\
1 & 6& -1\\
2 & -1& 2
\end{array}
\right],
B=
\left[\begin{array}{c}
7& -1\\
13& 6\\
5& 7
\end{array}
\right] 
$$

```python
#Generamos la matriz de coeficientes y la matriz de lado derecho
mtrxA = np.array([[1, 4, 1], [1, 6, -1], [2, -1, 2]])
mtrxB = np.array([[7, -1], [13, 6], [5, 7]])
print('Generamos la matriz de coeficientes y la matriz de lado derecho')
print('mtrxA')
pprint.pprint(mtrxA)
print('mtrxB')
pprint.pprint(mtrxB)
print('')

#Generamos las matricex Q y R con el módulo linalg de scipy:
mtrxQ, mtrxR = scipy.linalg.qr(mtrxA)
print('Generamos las matricex Q y R con el módulo linalg de scipy:')
print('mtrxQ')
pprint.pprint(mtrxQ)
print('mtrxR')
pprint.pprint(mtrxR)
print('')

#Generamos la matriz p, utilizando la traspuesta de Q, y aplicando el producto punto con B
mtrxP = np.dot(mtrxQ.T, mtrxB)
print('Generamos la matriz p, utilizando la traspuesta de Q, y aplicando el producto punto con B:')
print('mtrxQ.T')
pprint.pprint(mtrxQ.T)
print('mtrxB')
pprint.pprint(mtrxB)
print('mtrxP (np.dot(mtrxQ.T, mtrxB)):')
pprint.pprint(mtrxP)
print('')

#Generamos la matriz que contiene los vectores que resuelven el sistema A,B
#Para ello, obtenemos el producto punto entre la inversa de R y P
mtrxX = np.dot(scipy.linalg.inv(mtrxR), mtrxP) # Solve Rx=y 
print('Generamos la matriz que contiene los vectores que resuelven el sistema A,B')
print('Para ello, obtenemos el producto punto entre la inversa de R y P: ')
print('scipy.linalg.inv(mtrxR):')
pprint.pprint(scipy.linalg.inv(mtrxR))
print('mtrxP:')
pprint.pprint(mtrxP)
print('mtrxX np.dot(scipy.linalg.inv(mtrxR), mtrxP):')
pprint.pprint(mtrxX)
print('')

#Verificamos que al multiplicar AX, obtengamos B
print('Verificamos que al multiplicar AX, obtengamos B')
print('mtrxA:')
pprint.pprint(mtrxA)
print('mtrxX:')
pprint.pprint(mtrxX)
print('mtrxA@mtrxX:')
pprint.pprint(mtrxA@mtrxX)
print('mtrxB:')
pprint.pprint(mtrxB)
```
