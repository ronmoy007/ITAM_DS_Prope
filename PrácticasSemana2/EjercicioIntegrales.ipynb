Ejercicio: aproximar el valor de la integral $\displaystyle \int_0^{\pi}sin(x)dx = 2$ con reglas compuestas del rectángulo y trapecio con un número de subintervalos $n \in \{1,10,20,30,40 \}$ y hacer gráfica de la forma error relativo vs $n$ ($n$ en el eje horizontal)



```python
#Importamos
import numpy as np
import math
import matplotlib.pyplot as plt
import mjmaLibMat as LM
```

```python
#Funciones auxiliares

def ObtenerRsltYErrXRect(func, a, b, n, nbrResultadoReal):
    
    nbrRsltIntg=LM.IntegrarXRect(func, a, b, n)
    nbrErrorRel=LM.CalcErrorRelat(nbrRsltIntg, nbrResultadoReal)
    arrTmp=[n, nbrRsltIntg, nbrErrorRel]
    print("{} Intervalo(s) resultado:{} error relativo:{} ".format(n, nbrRsltIntg, nbrErrorRel))

    return arrTmp

def ObtenerRsltYErrXTrap(func, a, b, n, nbrResultadoReal):
    
    nbrRsltIntg=LM.IntegrarXTrap(func, a, b, n)
    nbrErrorRel=LM.CalcErrorRelat(nbrRsltIntg, nbrResultadoReal)
    arrTmp=[n, nbrRsltIntg, nbrErrorRel]
    print("{} Intervalo(s) resultado:{} error relativo:{} ".format(n, nbrRsltIntg, nbrErrorRel))

    return arrTmp

def ObtenerRsltYErrXGaussLeg(func, a, b, n, nbrResultadoReal):
    
    nbrRsltIntg=LM.IntegrarXGaussLeg(func, a, b, n)
    nbrErrorRel=LM.CalcErrorRelat(nbrRsltIntg, nbrResultadoReal)
    arrTmp=[n, nbrRsltIntg, nbrErrorRel]
    print("{} Intervalo(s) resultado:{} error relativo:{} ".format(n, nbrRsltIntg, nbrErrorRel))

    return arrTmp

def GraficarErroresRelatXInter(arrayRsltYErr):
    
    #Convertimos el array nativo a un array tipo numpy, para poder extraer fácilmente las columnas que necesitamos
    npArray=np.asarray(arrayRsltYErr)

    #Obtenemos de todas las filas, la primer columna (intervalos) y la tercer columna (errores relativos)
    arrayIntervalos=npArray[:,0]
    arrayErroresRel=npArray[:,2]

    #Puesto que el rango de errores relativos es muy amplio, usamos una escala logarítmica en el eje y
    plt.semilogy(arrayIntervalos,arrayErroresRel,'o-')
    plt.xlabel('Intervalos')
    plt.ylabel('Error relativo')
    plt.legend(('Error'),loc = 0)
    plt.grid(True)
    plt.show()

    return
```

```python
#Inicializamos las variables principales que se usarán a lo largo del código
nbrRangoInicio=0
nbrRangoFin=math.pi
nbrResultadoReal=2

#Arrays nativos
arrTmp=[]
arrRsltYErrXRect=[]
arrRsltYErrXTrap=[]
arrRsltYErrXGuass=[]
```

```python
#Flujo Integración por rectángulos
print('Integración por rectángulos')

#Hacemos el proceso para los distintos intervalos: 1, 10, 20, 30, 40
nbrIntervalo=1
arrTmp=ObtenerRsltYErrXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXRect.append(arrTmp)
LM.GraficarIntXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=10
arrTmp=ObtenerRsltYErrXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXRect.append(arrTmp)
LM.GraficarIntXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=20
arrTmp=ObtenerRsltYErrXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXRect.append(arrTmp)
LM.GraficarIntXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=30
arrTmp=ObtenerRsltYErrXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXRect.append(arrTmp)
LM.GraficarIntXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=40
arrTmp=ObtenerRsltYErrXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXRect.append(arrTmp)
LM.GraficarIntXRect(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

GraficarErroresRelatXInter(arrRsltYErrXRect)
```

```python
#Flujo Integración por trapezoide
print('Integración por trapezoide')

#Hacemos el proceso para los distintos intervalos: 1, 10, 20, 30, 40
nbrIntervalo=1
arrTmp=ObtenerRsltYErrXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXTrap.append(arrTmp)
LM.GraficarIntXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=10
arrTmp=ObtenerRsltYErrXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXTrap.append(arrTmp)
LM.GraficarIntXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=20
arrTmp=ObtenerRsltYErrXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXTrap.append(arrTmp)
LM.GraficarIntXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=30
arrTmp=ObtenerRsltYErrXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXTrap.append(arrTmp)
LM.GraficarIntXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

nbrIntervalo=40
arrTmp=ObtenerRsltYErrXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo, nbrResultadoReal)
arrRsltYErrXTrap.append(arrTmp)
LM.GraficarIntXTrap(np.sin, nbrRangoInicio, nbrRangoFin, nbrIntervalo)

GraficarErroresRelatXInter(arrRsltYErrXTrap)
```

```python

#Comparamos los errores relativos de ambos métodos
#Convertimos el array nativo a un array tipo numpy, para poder extraer fácilmente las columnas que necesitamos
npArrayRect=np.asarray(arrRsltYErrXRect)
npArrayTrap=np.asarray(arrRsltYErrXTrap)

#Obtenemos de todas las filas, la primer columna (intervalos) y la tercer columna (errores relativos)
arrayIntervalosRect=npArrayRect[:,0]
arrayErroresRelRect=npArrayRect[:,2]

arrayIntervalosTrap=npArrayTrap[:,0]
arrayErroresRelTrap=npArrayTrap[:,2]

#Puesto que el rango de errores relativos es muy amplio, usamos una escala logarítmica en el eje y
plt.semilogy(arrayIntervalosRect,arrayErroresRelRect,'o-')
plt.semilogy(arrayIntervalosTrap,arrayErroresRelTrap,'^-')
plt.xlabel('Intervalos')
plt.ylabel('Error relativo')
plt.legend(('Errores relativos rect','Errores relativos trap'),loc = 0)
plt.grid(True)
```

$$(2\pi\sigma^2)^{-\frac{1}{2}}\displaystyle \int_{-\infty}^\infty t^2e^{\frac{-(t-\mu)^2}{2\sigma^2}}dt$$
donde: $\sigma=0.25, \mu=0.15$ cuyos valores respectivamente son: $0.15, 0.085$ con cuadratura de Gauss-Hermite. Para lo anterior, realizar cambio de variable $x=\frac{t-\mu}{\sqrt{2\sigma^2}}, dt=\sqrt{2\sigma^2}dx$.

```python

#Declaramos e inicializamos variables
pi=math.pi
s=.25
m=.15
nbrConstante1=math.sqrt(2*s**2)   #0.3535533905932738
nbrConstante2=((2*pi*(s**2))**-.5)#1.5957691216057308

#Declaramos funciones
def funcTSimple(x):
    #Esta fórmula se obtiene de despejar t para expresarla en términos de s, m y x
    return m+(nbrConstante1*x)

def funcTCuadrada(x):
    #Esta fórmula se obtiene de despejar t para expresarla en términos de s, m y x
    return (m+(nbrConstante1*x))**2

###### Flujo principal
print('Ejercicio 1 Gauss hermite')

#Iteramos de 0 a 5 para mostrar los resultados de Gauss-Hermite en cada orden
for n in range(6):
    
    #Calculamos la integral con Gauss-Hermite y multiplicamos por las constantes que se presentan en el ejercicio
    nbrResultado=nbrConstante2*LM.IntegrarXGaussHer(n,funcTSimple)*nbrConstante1
    print('n: ',n, ' nbrResultado: ',nbrResultado)

print('')   
print('Ejercicio 2 Gauss hermite')

#Iteramos de 0 a 5 para mostrar los resultados de Gauss-Hermite en cada orden
for n in range(6):
    
    #Calculamos la integral con Gauss-Hermite y multiplicamos por las constantes que se presentan en el ejercicio
    nbrResultado=nbrConstante2*LM.IntegrarXGaussHer(n,funcTCuadrada)*nbrConstante1
    print('n: ',n, ' nbrResultado: ',nbrResultado)
```
